```python
#put image from image library here
live_test=Image.open("/Users/manorithakur/Desktop/Screenshot 2025-11-30 at 4.12.13 PM.png")
#code will now convert the image into grayscale(black and white) because it 
#is easier to process due to shade ranging from (0,255)
live_test=live_test.convert('L')
#reduces size of image for faster processing while maintaining quality using Lanczos
live_test=live_test.resize((28, 28), Image.LANCZOS)
#makes image into a a list of pixels 
x_demo = list(live_test.getdata())
```

```python
#load dataframes that allow the CNN to learn and test on
train=pd.read_csv('/Users/manorithakur/Downloads/mnist_train.csv')
test=pd.read_csv('/Users/manorithakur/Downloads/mnist_test.csv')

#Normalizes values in the X dataset and reshapes them to fit the model requirements
#Normalize: Divide all pixel values(0-225)
#Reshaping is required so that the kernel or pattern detector can scan the 28x28 array of pixels
def normalizer_and_reshaper_X(train_test):
    #normalizes dataset
    changedata = train_test.iloc[: , 1:].values/255.0
    #shapes dataset
    changedata = changedata.reshape(-1, 28, 28, 1)
    return changedata

#Creates an answer key in training and testing for telling 
#the results of getting a question correct or wrong
def label_Y(train_test):
    #retrieves only information in the column named label with answers to the training and testing
    labels = train_test['label'].values
    return labels

#demo is for showing what the AI predicted the image number is
x_demo = np.array(x_demo)/255.0
x_demo = x_demo.reshape(-1, 28, 28, 1)

#activates functions
x_train = normalizer_and_reshaper_X(train)
x_test = normalizer_and_reshaper_X(test)
y_train = label_Y(train)
y_test = label_Y(test)
```

```python
#model is the brains of the AI; the one that makes decisions on providing an answer 
#to each instance
#Note: Conv2D's 2nd value is 5 because the kernel is 5x5
#and in the second instance of Conv2D the kernel is 3x3
model=tf.keras.Sequential([
    tf.keras.layers.Conv2D(50, 5, activation='relu', input_shape=(28,28,1)), 
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)), 
    tf.keras.layers.Conv2D(50, 3, activation='relu'), 
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)), 
    tf.keras.layers.Flatten(), 
    tf.keras.layers.Dense(10, activation='softmax')
])
model.summary()
```

```python
#sets up what the model is measuring performance on
#loss is used to determine accuracy of training
#sparse_categorical_crossentropy: loss function in which the label or answer is found on the array
#and the negative logarithm is found of that number.
#Example from previous test session:
    #Softmax outputs:array([[7.6626020e-04, 8.0559886e-04, 9.9149406e-01, 1.2535170e-03,
        #4.7692980e-04, 2.4963247e-09, 2.1296383e-10, 4.5808339e-03,
        #7.7698769e-06, 6.1500317e-04]], dtype=float32)
    #Answer or label:2
    #sparse_categorical_crossentropy: find index 2 in array and find -log(9.9149406e-01)
    #Loss=-log(9.9149406e-01) solved out
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#model learns and improves upon mistakes
#epochs means the number of times the model repeats the learning process
print("Train")
model.fit(x_train, y_train, epochs=3)
```

```python
#model uses what it has learned and tests itself for a final accuracy check
print("Test")
model.fit(x_test, y_test, epochs=1)
```

```python
#makes a prediction on what the image's number is using the training, testing, 
#and model knowledge
print("Demo")
np.argmax(model.predict(x_demo))
```
