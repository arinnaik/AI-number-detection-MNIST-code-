```python
import pandas as pd
import numpy as np
import tensorflow as tf
print(tf.__version__)
print(np.__version__)
print(pd.__version__)
train=pd.read_csv('/Users/manorithakur/Downloads/mnist_train.csv')
test=pd.read_csv('/Users/manorithakur/Downloads/mnist_test.csv')
model=tf.keras.Sequential([
    tf.keras.layers.Conv2D(50, 5, activation='relu', input_shape=(28,28,1)), tf.keras.layers.MaxPooling2D(pool_size=(2,2)), tf.keras.layers.Conv2D(50, 3, activation='relu'), tf.keras.layers.MaxPooling2D(pool_size=(2,2)), tf.keras.layers.Flatten(), tf.keras.layers.Dense(10)
])
model.summary()
train=train.sample(frac=1).reset_index(drop=True)
mocktest=train.iloc[0:10000]
y_train = train['label']
x_train = train.iloc[: , 1:]/255
y_test = test['label']
x_test = test.iloc[: , 1:]/255
for i in range(60000):
    label=y_train.iloc[i]
    row=np.array(x_train.iloc[i])
    row=row.reshape(28, 28)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    model.fit(row, label, epochs=5)
for i in range(10000):
    row=np.array(x_test.iloc[i])
    row=row.reshape(28, 28)
    print(row)
```
